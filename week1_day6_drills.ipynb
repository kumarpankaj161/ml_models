```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Quick clean (same as Day 5)
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns=['Cabin'])

print('Ready! Shape:', df.shape)
print(df.columns.tolist())
```

```python
print(df['Pclass'].unique())
print(df['Pclass'].nunique())
```

```python
df.groupby('Pclass')['Fare'].mean().round(2)
```

```python
avg = df['Fare'].mean()
print(len(df[df['Fare'] > avg]))
```

```python
def fare_cat(f):
    if f < 15: return 'Low'
    elif f <= 50: return 'Mid'
    else: return 'High'
df['FareCategory'] = df['Fare'].apply(fare_cat)
print(df['FareCategory'] )
```
